```python
import os
import pandas as pd
from tqdm import tqdm
import numpy as np
from sklearn import linear_model
import cvxpy as cp
from threading import Thread

dataBasePath = r'D:\Local Database\tmp_data\ETF_Replication_data'

adjClose = pd.read_parquet(os.path.join(dataBasePath,'ADJCLOSE.parquet'))
stockInd = pd.read_parquet(os.path.join(dataBasePath,'ZXIND.parquet'))
record = pd.read_pickle(os.path.join(dataBasePath,'399303SZ_MEMBER.pkl'))
record.loc[:, 'S_CON_INDATE'] = pd.to_datetime(record.loc[:, 'S_CON_INDATE'], format='%Y%m%d')
record.loc[:, 'S_CON_OUTDATE'] = pd.to_datetime(record.loc[:, 'S_CON_OUTDATE'], format='%Y%m%d')
tradeDates = adjClose.index[adjClose.index >= record.loc[:,'S_CON_INDATE'].min()].tolist()
stockList = adjClose.columns.tolist()
indexClose = pd.read_excel(os.path.join(dataBasePath,'指数收盘价.xlsx'), index_col=0).loc[tradeDates,:]
indexClose = indexClose / indexClose.iloc[0, :]

adjClose = adjClose.loc[tradeDates, :]
mv = pd.read_parquet(os.path.join(dataBasePath,'FREEFLOATMV.parquet')).loc[tradeDates,:]

indexRet = indexClose.pct_change(1)
adjRet = adjClose.pct_change(1)

indexConst = pd.DataFrame(0, index=tradeDates, columns=stockList)
for _, rowValue in record.iterrows():
    stock = rowValue.loc['S_CON_WINDCODE']
    indate = rowValue.loc['S_CON_INDATE']
    outdate = rowValue.loc['S_CON_OUTDATE']
    if isinstance(outdate, pd._libs.tslibs.nattype.NaTType):
        indexConst.loc[indate:, stock] = 1
    else:
        indexConst.loc[indate:outdate, stock] = 1
```

```python
### tradeDates: list(pd.Timestamp), 记录交易日
### stockList: list(str), 记录股票代码
### adjClose/mv/adjRet/indexConst：pd.DataFrame(tradeDates, stockList)，其中indexConst为0/1矩阵
### indexClose/indexRet：pd.DataFrame(tradeDates, indexCode)，记录指数数据，初始为1。
```

### 最小化跟踪误差

```python
h = 243
alpha = 1e-6
optSimu = pd.Series(1, index=tradeDates, name='Optimized Simulation')
dailyWeight = pd.DataFrame(0., index=tradeDates, columns=stockList, dtype=float)

def get_stocks_by_lasso(X, y, alpha):
    lasso = linear_model.Lasso(alpha=alpha,fit_intercept=False, max_iter=int(1e6))
    lasso.fit(X, y)
    return lasso.coef_
    
def slove_by_cvxpy(X, y, n, isset_psd=False):
    if isset_psd:
        P = cp.psd_wrap(X.T @ X)
    else:
        P = X.T @ X
    q = -2 * y.T @ X
    G = -1 * np.identity(n)
    h = np.zeros(n)
    A = np.ones(n)
    b = 1
    w = cp.Variable(n)
    prob = cp.Problem(
        cp.Minimize(cp.quad_form(w, P) + q.T @ w),
        [G @ w <= h, A @ w == b])
    prob.solve(solver=cp.CVXOPT, kktsolver=cp.ROBUST_KKTSOLVER)
    return w.value

for thisDay in tqdm(tradeDates[h+1:]):
    preDay = tradeDates[tradeDates.index(thisDay) - 1]
    day = preDay

    retStock = adjRet.loc[:day, (indexConst.loc[day, :]==1)].iloc[-h:-1, :].dropna(how='any', axis=1)
    retStock = retStock.loc[:, ~(retStock == 0).all(axis=0)] # 剔除收益率全为0
    y = indexRet.loc[:day,'CN2303.CNI'].iloc[-h:-1].values
    
    res = get_stocks_by_lasso(retStock.values, y, alpha)
    lasso_res = pd.Series(res, index=retStock.columns.tolist())
    stockChosen = lasso_res.loc[lasso_res>0].index.tolist()
    n = len(stockChosen)
    X = adjRet.loc[:day, stockChosen].iloc[-h:-1, :].fillna(0).values
    try:
        w_solve = slove_by_cvxpy(X, y, n)
    except Exception as e:
        print(f'Retry... Error: {e}')
        w_solve = slove_by_cvxpy(X, y, n, True)

    weight = pd.Series(w_solve, index=stockChosen)

    weight = weight / weight.sum()
    dailyWeight.loc[preDay, weight.index.tolist()] = weight.values
    holdNum = optSimu.loc[preDay] * weight / adjClose.loc[preDay, weight.index]
    optSimu.loc[thisDay] = (holdNum * adjClose.loc[thisDay, weight.index]).sum()
optSimu.plot(legend=True)
indexClose['CN2303.CNI'].plot(legend=True)
```

### 分行业->高相关性(基础池)->最小化跟踪误差(优选池)->给定持仓比例

```python
h = 243
ratio = 0.3
alpha = 1e-7

finalSimu = pd.Series(1, index=tradeDates, name='Final Simulation')
dailyWeight = pd.DataFrame(0., index=tradeDates, columns=stockList, dtype=float)

def get_high_corr_by_ind(ratio, day, h):
    stocksCorr = adjRet.loc[:day, indexConst.loc[day, :]==1].iloc[-h:-1, :].apply(lambda x: np.correlate(x.values, indexRet.loc[:day, 'CN2303.CNI'].iloc[-h:-1].values), axis=0).iloc[0,:]
    stocksChosen = []
    for ind in allInds:
        conIndStocks = indexConst.loc[day, (stockInd.loc[day, :] == ind) & (indexConst.loc[day, :] == 1)].index.tolist()
        n = len(conIndStocks)
        stocksIndChosen = stocksCorr.loc[conIndStocks].sort_values(ascending=False).index[: int(ratio*n)].tolist()
        stocksChosen.extend(stocksIndChosen)
    return stocksChosen

def get_stocks_by_lasso(stockRaw, day, alpha, h):
    retStock = adjRet.loc[:day, stockRaw].iloc[-h:-1, :].dropna(how='any', axis=1)
    retStock = retStock.loc[:, ~(retStock == 0).all(axis=0)] # 剔除收益率全为0
    X = retStock.values
    y = indexRet.loc[:day,'CN2303.CNI'].iloc[-h:-1].values
    
    lasso = linear_model.Lasso(alpha=alpha,fit_intercept=False, max_iter=int(1e6))
    lasso.fit(X, y)
    
    lasso_res = pd.Series(lasso.coef_, index=retStock.columns.tolist())
    stockChosen = lasso_res.loc[lasso_res>0].index.tolist()
    return stockChosen
    
def slove_by_cvxpy(X, y, isset_psd=False):
    n = X.shape[1]
    if isset_psd:
        P = cp.psd_wrap(X.T @ X)
    else:
        P = X.T @ X
    q = -2 * y.T @ X
    G = -1 * np.identity(n)
    h = np.zeros(n)
    A = np.ones(n)
    b = 1
    w = cp.Variable(n)
    prob = cp.Problem(
        cp.Minimize(cp.quad_form(w, P) + q.T @ w),
        [G @ w <= h, A @ w == b])
    prob.solve(solver=cp.CVXOPT, kktsolver=cp.ROBUST_KKTSOLVER)
    return w.value

def get_stocks_weight(stockChosen, day, h):
    X = adjRet.loc[:day, stockChosen].iloc[-h:-1, :].fillna(0).values
    y = indexRet.loc[:day,'CN2303.CNI'].iloc[-h:-1].values
    try:
        w_solve = slove_by_cvxpy(X, y)
    except Exception as e:
        print(f'Retry... Error: {e}')
        w_solve = slove_by_cvxpy(X, y, True)
    weight = pd.Series(w_solve, index=stockChosen)
    return weight / weight.sum()


def full_process(idx, all_weights):
    dates = tradeDates[h+1:][idx::4]
    for thisDay in dates:
        preDay = tradeDates[tradeDates.index(thisDay) - 1]
        stockHighCorr = get_high_corr_by_ind(ratio, preDay, h)
        stockFiltered = get_stocks_by_lasso(stockHighCorr, preDay, alpha, h)
        weight = get_stocks_weight(stockFiltered, preDay, h)
        all_weights[thisDay] = weight
        print(thisDay)

    
all_weights = {}
threads = []

for i in range(4):
    t = Thread(target=full_process, args=(i, all_weights))
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()
```

```python
for thisDay in tqdm(tradeDates[h+1:]):
    weight = all_weights[thisDay]
    preDay = tradeDates[tradeDates.index(thisDay) - 1]
    holdNum = finalSimu.loc[preDay] * weight / adjClose.loc[preDay, weight.index]
    finalSimu.loc[thisDay] = (holdNum * adjClose.loc[thisDay, weight.index]).sum()

(finalSimu.iloc[h:]/finalSimu.iloc[h]).plot(legend=True)
(indexClose['CN2303.CNI'].iloc[h:]/indexClose['CN2303.CNI'].iloc[h]).plot(legend=True)
```

```python
def get_TD(ETFIndex, indexValue):
    ret = pd.concat([ETFIndex.pct_change(1), indexValue.pct_change(1)],axis=1).dropna(how='any')
    ret['TD'] = ret[ETFIndex.name] - ret[indexValue.name]
    TD = ret['TD'].abs().rolling(243).mean()
    return TD

def get_TE(ETFIndex, indexValue):
    ret = pd.concat([ETFIndex.pct_change(1), indexValue.pct_change(1)],axis=1).dropna(how='any')
    ret['TD'] = ret[ETFIndex.name] - ret[indexValue.name]
    TE = ret['TD'].rolling(243).apply(lambda x: ((x - x.mean())**2).mean()) * np.sqrt(243)
    return TE

get_TE(finalSimu.iloc[h:], indexClose['CN2303.CNI'].iloc[h:]).dropna().plot()
```
